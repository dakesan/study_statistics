統計のウソの例ならば、反ワク関連に関してネタが尽きない。

https://square.umin.ac.jp/~massie-tmd/3rdjab_excess_mortality.html

このページではワクチン接種が推進された2022年二月三月で志望者が増加していたため、ワクチン接種が危険なものであると結論づけている。

```python
df_death = pl.read_csv("https://covid19.mhlw.go.jp/public/opendata/deaths_cumulative_daily.csv")
df_death = df_death.with_column(
    pl.col("Date").str.strptime(
        pl.Date, fmt="%Y/%m/%d"
    ).cast(pl.Datetime)
)
df_death
```

```python
df_vaccine = pl.read_csv("data/summary_by_date.csv")
df_vaccine = df_vaccine.with_column(
    pl.col("date").str.strptime(
        pl.Date, fmt="%Y-%m-%d"
    ).cast(pl.Datetime)
)
df_vaccine
```

```python
df_severe = pl.read_csv("https://covid19.mhlw.go.jp/public/opendata/severe_cases_daily.csv")
df_severe = df_severe.with_column(
    pl.col("Date").str.strptime(
        pl.Date, fmt="%Y/%m/%d"
    ).cast(pl.Datetime)
)
df_severe
```

```python
fig, ax = plt.subplots()
fig.set_figwidth(15)

sns.lineplot(x="date", y="count_first_shot_general", data=df_vaccine, ax = ax,
             legend="brief", label="1st")
sns.lineplot(x="date", y="count_second_shot_general", data=df_vaccine, ax = ax,
             legend="brief", label="2nd")
sns.lineplot(x="date", y="count_third_shot_general", data=df_vaccine, ax = ax,
             legend="brief", label="3rd")
sns.lineplot(x="date", y="count_fourth_shot_general", data=df_vaccine, ax = ax,
             legend="brief", label="4th")
ax2 = ax.twinx()
sns.lineplot(x="Date", y="ALL", data=df_death, ax=ax2,
             legend="brief", label="death")
```

```python
fig, ax = plt.subplots()
fig.set_figwidth(15)

sns.lineplot(x="Date", y="ALL", data=df_death, ax=ax,
             legend="brief", label="death")
ax2 = ax.twinx()
sns.lineplot(x="Date", y="ALL", data=df_severe, ax=ax2,
             legend="brief", label="severe cases")
plt.legend(loc="upper left")
```

```python
fig, ax = plt.subplots()
fig.set_figwidth(15)

sns.lineplot(x="date", y="count_first_shot_general", data=df_vaccine, ax = ax,
             legend="brief", label="1st")
sns.lineplot(x="date", y="count_second_shot_general", data=df_vaccine, ax = ax,
             legend="brief", label="2nd")
sns.lineplot(x="date", y="count_third_shot_general", data=df_vaccine, ax = ax,
             legend="brief", label="3rd")
sns.lineplot(x="date", y="count_fourth_shot_general", data=df_vaccine, ax = ax,
             legend="brief", label="4th")
ax2 = ax.twinx()
sns.lineplot(x="Date", y="ALL", data=df_severe, ax=ax2,
             legend="brief", label="severe cases")
plt.legend(loc="upper left")
```

確かにワクチン接種時に死亡数は増加しているが、同時期には感染の波が来ており、重症者数も増加していることが分かる。ワクチン接種と死亡数の相関が見られるが、そこには因果関係が無く、疑似相関の典型であると考えられる。

明らかにワクチン接種数とCOVID-19症例の死亡であれば後者の方が死亡数には寄与度が大きいはずであり、相関の裏にある、見せられていないデータまで探ることで正しい解釈が可能になる。
